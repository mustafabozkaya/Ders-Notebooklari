## Kütüphaneler

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["xtick.labelsize"] = 14
plt.rcParams["ytick.labelsize"] = 14

plt.rcParams["axes.labelsize"] = 16
```

```python
data = sns.load_dataset("penguins")
data
```

```python
data[data["sex"] == "Male"]["flipper_length_mm"]
```

### Seaborn Olmadan Histogram

```python
?plt.hist
```

```python
data_male = data[data["sex"] == "Male"]["flipper_length_mm"]
data_female = data[data["sex"] == "Female"]["flipper_length_mm"]

plt.hist(x=data_female, label = "Female", alpha=0.5)
plt.hist(x=data_male, label = "Male", alpha = 0.5)

plt.legend()
plt.show()
```

### Seaborn ile Histogram

```python
sns.histplot(data=data, x="flipper_length_mm", hue = "sex", alpha=0.5)
plt.show()
```

## Seaborn ile Scatterplot

```python
data
```

```python
sns.scatterplot(
    data=data,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    style="sex",
    size="body_mass_g"
)

plt.title("bill_length_mm vs bill_depth_mm")
plt.show()
```

## Seaborn ile Barplot

```python
data
```

```python
grouped_data = data.groupby(["species", "sex"])["body_mass_g"].mean().reset_index()
grouped_data
```

```python
sns.barplot(data = grouped_data,
            x="species", y="body_mass_g",
            hue="sex")
plt.show()
```

## Subplot Oluşturma

### plt.subplots

```python
plt.subplots(2, 2, figsize=(12, 10))
```

#### Soru: Türlerin (species), body_mass_g özelliğine bakarak bir boxplot çiziniz ve subplot içine ekleyiniz.
x = species, y = body_mass_g

#### Soru: bill_length_mm sütununu bir histogram olarak subplot içinde plot ediniz. "species"'e göre ayrılsın.
hue=...

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

sns.scatterplot(
    data=data,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    ax=axes[0, 0]
)

sns.boxplot(
    data=data,
    x = "species",
    y = "body_mass_g",
    ax=axes[0, 1]
)

sns.histplot(
    data=data,
    x = "bill_length_mm",
    ax=axes[1, 0],
    hue="species"
)

sns.barplot(data = data,
            x="species", y="body_mass_g",ax=axes[1, 1],
            hue="sex")

axes[0, 0].set_title("bill_length_mm vs bill_depth_mm")
axes[0, 1].set_title("Spicies & BodyMassG Boxplot")
axes[1,0].set_title("bill_length_mm dist. Histplot")
axes[1, 1].set_title("Spicies & BodyMassG Barplot")
```

## Colormap ekleme

```python
grouped_data = data.groupby(["species"])["body_mass_g"].mean()
grouped_data
```

```python
grouped_data.index
```

```python
cmap = plt.get_cmap("viridis") # the other cmap
cmap
```

```python
norm = plt.Normalize(grouped_data.min(), grouped_data.max())
```

```python
colors = cmap(norm(grouped_data.values))
colors
```

```python
fig, ax = plt.subplots(figsize = (12, 9))

sns.barplot(x=grouped_data.index,
            y=grouped_data.values,
            palette=colors)

ax.set_ylabel("Average body_mass_g")

colorax = fig.add_axes([0.95, 0.11, 0.06, 0.77]) # add colormap axis 
"""
rect = l, b, w, h
fig = plt.figure()
fig.add_axes(rect)
"""
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
fig.colorbar(sm, cax=colorax)

plt.show()
```
