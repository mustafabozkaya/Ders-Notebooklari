`plt.plot()` fonksiyonu, Matplotlib kütüphanesinde temel çizgi grafiği çizmek için kullanılır. İki liste veya dizi arasındaki ilişkiyi göstermek için x ve y eksenlerinde değerleri birleştiren düz çizgiler çizer.

Fonksiyonun kullanımı oldukça basittir. İki eşit uzunluktaki liste veya dizi alır ve bunları x ve y eksenlerinde değerler olarak kullanarak bir çizgi grafiği oluşturur.

Örnek olarak, iki liste arasındaki ilişkiyi gösteren basit bir çizgi grafiği oluşturalım:

```python
# modellediğimiz ilişki - fonksiyon = f(x) = 2x
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (6, 4))

plt.plot(x, y)

plt.show()
```

### Soru: -3x grafiğini gosteren çizgi grafiğini (plt.plot) çiziniz.

```python
# modellediğimiz ilişki - fonksiyon = f(x) = -3x
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, -3, -6, -9, -12, -15, -18]

plt.figure(figsize = (6, 4))

plt.plot(x, y)

plt.show()
```

## plt.scatter

```python
?plt.scatter
```

`plt.scatter()` fonksiyonu, Matplotlib kütüphanesinde nokta grafiği (scatter plot) çizmek için kullanılır. Nokta grafiği, iki değişken arasındaki ilişkiyi göstermek için x ve y eksenlerinde noktaların dağılımını çizerek görselleştirir. Bu tür grafikler, iki değişken arasında doğru veya ters ilişkiler gibi eğilimleri ve kümelenmeleri görmeye yardımcı olabilir.

`plt.scatter()` fonksiyonu, `plt.plot()` fonksiyonuna benzer şekilde çalışır, ancak noktaları birleştiren düz çizgiler yerine veri noktalarının dağılımını gösterir. İki eşit uzunluktaki liste veya dizi alır ve bunları x ve y eksenlerinde değerler olarak kullanarak bir nokta grafiği oluşturur.

```python
# modellediğimiz ilişki - fonksiyon = f(x) = 2x
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (6, 4))

# Aynı anda plt.plot() ve plt.scatter() fonksiyonlarını kullanarak çizgi grafiği (plt.plot)
# ve nokta grafiğini (scatter plot) aynı figure üzerinde çizdirebilirsiniz:
plt.scatter(x, y, color = "red")
plt.plot(x, y, color="blue")

plt.show()
```

### Soru: Numpy kullanarak cos(x) ve sin(x) grafiklerini çizdirelim.
 * cos(x) plt.scatter, sin(x) plt.plot
 * x = np.arange(300) / 10

```python
x = np.arange(300)/10
y1 = np.sin(x)
y2 = np.cos(x)

plt.scatter(x, y2, color="red", label = "Cosx")
plt.plot(x, y1, color = "green", label = "Sinx")

plt.legend(loc="upper right")
plt.show()
```

## plt.plot ve plt.scatter Parametreleri

### plt.plot

`color`: çizginin rengini belirler

`linestyle`: çizgi stilini belirler

`marker`: noktalara denk gelen kesişimleri gösterir. "o,s,+" olabilir.

`markersize`: marker boyutudur 2,5 olabilir

`alpha`: saydamlık - opaklık. 1=opak, 0=saydam

`label`: grafiğin etiketi

```python
plt.plot(x, y1, color = "magenta", label = "Sinx",
         linestyle="--", )

plt.legend(loc="lower left")
plt.show()
```

```python
x = np.arange(10)
y = 2*x

plt.figure(figsize = (6, 4))

plt.plot(x, y,":", color="blue", marker="s", 
         markersize=10, alpha=0.5,markerfacecolor="yellow")

plt.show()
```

### plt.scatter parametreleri

`color`: renk belirtir

`label`: grafiğin etiketi

`s`: noktaların büyüklüğü

`edgecolors`: noktaları saran renk.

`alpha`: saydamlık ve opaklık

```python
x = np.arange(100) / 10
plt.scatter(x, np.cos(x),color="orange", label = "Cosx",
            s=20, edgecolors="green", linewidths=1,
            alpha=0.5)

plt.legend()
plt.show()
```

## Parametreler ve Özet
Parametrelerin birçoğu hem `plt.scatter()` hem de `plt.plot()` fonksiyonlarında kullanılabilir, ancak fonksiyonların kendine özgü özellikleri vardır. Örneğin, `plt.scatter()` fonksiyonu, veri noktalarının boyutunu ve renklerini değiştirmenize olanak tanırken, `plt.plot()` fonksiyonu, çizgi grafiği çizmek için daha fazla seçenek sunar.

İki fonksiyonun farklı kullanım amaçları vardır. `plt.scatter()` daha çok iki değişken arasındaki ilişkiyi ve dağılımı göstermek için kullanılırken, `plt.plot()` genellikle zaman serisi verileri veya sürekli fonksiyonlar gibi bağlantılı verileri göstermek için kullanılır.
